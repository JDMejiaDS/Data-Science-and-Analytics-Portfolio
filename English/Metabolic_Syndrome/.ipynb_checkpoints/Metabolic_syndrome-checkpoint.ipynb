The Kolmogorov-Smirnov test was conducted to evaluate whether the ages of healthy and sick individuals follow a normal distribution. The p-values obtained were extremely low (p-value = 0.0) in both groups, indicating a significant discrepancy with the normality hypothesis. Therefore, the normality assumptions required by parametric tests are not met.

Additionally, a Levene's test was performed to assess the homogeneity of variances between the groups. The result showed a significantly low p-value (p-value = 0.000296), indicating a significant difference in the variances of ages between the healthy and sick groups. This contradicts the assumption of homogeneity of variances required for parametric tests.

Given that the data do not follow a normal distribution and the variances are not homogeneous, it justifies the use of a non-parametric approach for comparing ages between the groups. Therefore, a non-parametric test, such as the Mann-Whitney U test, will be performed, which is robust in the presence of such deviations from parametric assumptions.

```python
from scipy.stats import kstest, levene
```

```python
H_age = df[df["MetabolicSyndrome"]==0]["Age"]
S_age = df[df["MetabolicSyndrome"]==1]["Age"]
stat_h, healthy_p_value = kstest(H_age, "norm")
stat_s, sick_p_value = kstest(S_age, "norm")
print(f"The p value for Kolmogorov test for healthy individuals is: {healthy_p_value}")
print(f"The p value for Kolmogorov test for sick individuals is:  {sick_p_value}")
stat_levene, p_value_levene = levene(H_age, S_age)
print(f'Levene test p-value is :{p_value_levene}')
```

A Mann-Whitney U test was conducted to compare the ages between healthy individuals and those with metabolic syndrome. The test results indicate a significant difference in the age distributions between the two groups (p-value < 0.05).

The p-value obtained was extremely low, with a p-value equal to 3.04 x 10^-35. This indicates substantial evidence against the null hypothesis that the medians of ages in both groups are equal. Therefore, we can conclude that there is a statistically significant difference in ages between healthy individuals and those with metabolic syndrome.

With all the graphical and statistical evidence presented, we can infer that there is a directly proportional trend between age and the likelihood of having metabolic syndrome. As seen in the previous distplot, there is a concentration of individuals with metabolic syndrome around older ages, whereas younger individuals seem to have a much lower likelihood of having it.

```python
from scipy.stats import mannwhitneyu
MWU_stati, p_value_age = mannwhitneyu(S_age, H_age)
print(f'The p-value asociated to U de Mann - Whitney test is :{p_value_age}')
```

As noted earlier, there are some missing data for four of the variables studied, as shown in the section immediately preceding this text. Before proceeding further with the analysis, it is necessary to perform data imputation on the dataset to capture as much information as possible by estimating those missing values. To accomplish this task, the Scikit-learn library will be implemented, using the K-nearest neighbors imputation technique.

The K-nearest neighbors technique allows us to handle missing numerical data straightforwardly. However, this only works for numerical data. Imputing categorical values is a bit more complicated. This time, I will handle these categorical fields by assigning a value to each one based on the probability of occurrence of all values already present in this dimension. I will do this as follows:

```python
from sklearn.impute import KNNImputer
imput = KNNImputer(n_neighbors= 6)
df_ip = df.copy()
faltantes_num = df[["Income", "WaistCirc", "BMI"]]
imputed = imput.fit_transform(faltantes_num)
```

```python
imputed_val = pd.DataFrame(imputed, columns=["Income", "WaistCirc", "BMI"])
imputed_val.head()
```

```python
df_ip = df_ip.drop(["Income","WaistCirc","BMI"], axis = 1)
```

```python
df_ip.reset_index(inplace=True)
```

```python
df_clas = pd.concat([df_ip,imputed_val], axis=1)
df_clas.head()
```

```python
df_clas.isna().sum()
```

```python
class_prob = df_clas["Marital"].value_counts(normalize=True)
```

```python
def imput_marital(data):
    if pd.isna(data):
        data = np.random.choice(class_prob.index, p=class_prob.values)
    return data
```

```python
df_clas["Marital"] = df_clas["Marital"].apply(lambda x: imput_marital(x))
```

```python
df_clas.info()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
meta_synd_PCA = pca.fit_transform(df_NN)
meta_synd = pd.DataFrame(meta_synd_PCA)
plt.figure(figsize=(8,6))
sns.scatterplot(x = meta_synd[0], y =meta_synd[1], hue =df_NN['MetabolicSyndrome'], palette="viridis")
plt.xlabel('1st PC')
plt.ylabel('2nd PC')
plt.title("PCA Metabolic syndrome")
plt.legend(labels = ["Healthy","Sick"])
plt.show()
```

After using PCA, we identified 4 distinct clusters that are conspicuously delineated. Since we don't know which features might be grouping these observations, we can perform a clustering procedure and analyze the results to understand which features contribute to these particular patterns. I will conduct this analysis as follows:

At first glance, there are 4 clusters of points. However, I will use a metric called Within Cluster Sum of Squares (WCSS) to evaluate the quality of the clusters based on the similarity of the points within each cluster. This forms the basis of the Elbow method, which visualizes a plot describing the error within clusters as a function of the number of clusters formed. Using this technique, we can determine the optimal number of clusters to form.

```python
from sklearn.cluster import KMeans
WCSS = []
for i in range(1,16):
    kmeans = KMeans(i, n_init="auto")
    kmeans.fit(df_NN)
    WCSS_inertia= kmeans.inertia_
    WCSS.append(WCSS_inertia)
```

As seen in the figure below this paragraph, the intra-cluster error decreases sharply as the number of clusters increases. To find the ideal number of clusters, it is best to locate the point where the error rate sharply decreases, typically referred to as the 'elbow point'. This point usually corresponds to the optimal number of clusters. However, determining this isn't always a precise value but rather falls within a range of values. Domain knowledge or prior data visualization can aid in making this choice. Given the scatter plot created with PCA and the previous figure, it can be concluded that indeed, the optimal number of clusters is 4, both from visual inspection and the range of values shown in the WCSS plot against the number of clusters.

```python
num = range(1,16)
plt.plot(num,WCSS)
plt.title("Elbow for number of clusters selection")
plt.ylabel("Intra-clusters error")
plt.xlabel("Num of clusters")
plt.show()
```

```python
kmeans = KMeans(4, n_init="auto")
kmeans.fit(df_NN)
df_clas["Cluster"] = kmeans.labels_
df_clas.head()
```

To conclude, I will create a neural network that allows me to classify based on the information extracted from the dataset. However, as shown in the bar plot at the beginning of the document, the target variable exhibits class imbalance: class 0 (healthy) has approximately twice as many observations as class 1 (ill). To address this, I will use a technique called SMOTE, which utilizes the K-nearest neighbors technique to generate synthetic samples, aiming to avoid overfitting that might occur with simple resampling with replacement (though this remains an option in specific scenarios). Continuing:

```python
from imblearn.over_sampling import SMOTE
X = df_NN.drop("MetabolicSyndrome", axis = 1)
y = df_NN["MetabolicSyndrome"]
X_resamp, y_resamp = SMOTE().fit_resample(X,y)
```

```python
y_resamp.value_counts()
```

Now that the imbalance of classes has been corrected, the neural network can be created

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report
tf.random.set_seed(96)
```

It is important to create three data splits: one for training, another for validation, and a third for testing. This is to closely monitor how the model is being trained and performing, as it provides insight into whether the model is truly effective or not. Below, I will generate these three partitions before proceeding with model generation.

```python
X_train, X_test, y_train, y_test = train_test_split(X_resamp, y_resamp, test_size=0.15, shuffle=True)
```

```python
tr_len = int(len(X_train)*0.85)
val_len = int(len(X_train)*0.15)
print(tr_len, val_len)
```

```python
X_tr = X_train[:tr_len]
X_val = X_train[tr_len:]
y_tr = y_train[:tr_len]
y_val = y_train[tr_len:]
print(X_tr.shape, X_val.shape, y_tr.shape,y_val.shape)
```

```python
print(X_test.shape, y_test.shape)
```

```python
modelo = Sequential([
    Dense(64, activation='relu', input_shape=(10,)), Dropout(0.4),
    Dense(32, activation='relu'), Dropout(0.3),
    Dense(1, activation='sigmoid')
])
modelo.compile(optimizer="adam" , loss='binary_crossentropy', metrics=['accuracy'])
modelo.fit(X_tr, y_tr,epochs=100, batch_size=32, verbose = 0, validation_data=(X_val, y_val))
```

After training is complete, the loss and accuracy curves are plotted for both the training and validation datasets. This visual confirmation helps ensure that overfitting is not occurring and that the model's applicability and performance on unseen datasets are optimized."

This approach will give you insights into how well your model generalizes to unseen data by examining its performance on both the training and validation sets

```python
plt.title("Pérdida por ciclo de entrenamiento")
plt.plot(modelo.history.history["loss"], label ="Pérdida", linestyle = "dashed", color = "red" )
plt.plot(modelo.history.history["val_loss"], label ="Pérdida validación", linestyle = "dashed", color = "teal" )
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.legend()
plt.show()
```

Based on the assessment of both the loss and accuracy curves, it appears that the model is performing well and not exhibiting signs of overfitting. However, considering the context of medical applications, typically a high level of accuracy is crucial for deployment in real-world scenarios. An accuracy below 95% might be considered inadequate for medical decision-making where precision and reliability are paramount.

If the model reaches an accuracy threshold of 95% or higher, it would likely be more suitable for deployment in real medical settings. This ensures that the predictions are consistently reliable and trustworthy, which is essential when making decisions related to patient health based on machine learning models.
